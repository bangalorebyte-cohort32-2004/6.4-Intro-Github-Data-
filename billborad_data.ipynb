```python
import pandas as pd

df =pd.read_csv('./assignments/billboard.csv')
df
```

```python
df = df.melt(id_vars=['year' ,'artist', 'track' ,'time' ,'date.entered'], var_name='week', value_name='rank')
df = df.dropna()
df
```

```python
df = df.reset_index()
df
```

```python
da = []
for i in range(0,5307):
    j = df['week'][i][2:]
    da.append(j)
da
```

```python
df['week'] = da
df
```

```python
df.sample(50)
```

```python
import datetime

da = []
for i in range(0,5307):
    j = df['date.entered'][i]
    w = int(df['week'][i])
    w = (w -1) * 7
    t = datetime.datetime.strptime(j, "%Y-%m-%d")
    t = t + datetime.timedelta(days=w)
    t = t.strftime('%Y-%m-%d')
    da.append(t)
da 
```

```python
da[393]
```

```python
df['date.entered'] = da
df = df.rename(columns={'date.entered': 'date'})
df
```

```python
df['date'][393]
```

```python
df = df.sort_values(by=['artist','date'])
df
```
