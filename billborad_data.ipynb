```python
import pandas as pd

df =pd.read_csv('./assignments/billboard.csv')
df
```

```python
df = df.melt(id_vars=['year' ,'artist', 'track' ,'time' ,'date.entered'], var_name='week', value_name='rank')
df = df.dropna()
df = df.reset_index()
df
```

```python
df[['c','week1']] = df['week'].str.split('k',expand=True)
df = df.drop(columns=['c','week'])
df.rename(columns={'week1':'week'},inplace=True)
df['week'] = df['week'].astype(int)
df
```

```python
df['date'] = pd.to_datetime(df['date.entered'])
df = df.drop(columns='date.entered')
```

```python
import numpy as np

x = (((df['week']) - 1) * 7)
df['date'] = df['date'] + pd.to_timedelta(np.ceil(x), unit="D")

df
```

```python
df = df.sort_values(by=['artist','date'])
df = df.reset_index()
df = df.rename(columns={'level_0': 'id'})
del df['index']
df = df.assign(id=(df['track']).astype('category').cat.codes)
df = df[['id','year','artist','track','time','date','week','rank']]
```

```python
df_s = df.copy()
del df_s['year']
del df_s['date']
del df_s['week']
del df_s['rank']
df_s.drop_duplicates(inplace=True)
```

```python
df_d = df.copy()
del df_d['year']
del df_d['artist']
del df_d['track']
del df_d['time']
del df_d['week']
```
