```python
import pandas as pd

df =pd.read_csv('./assignments/country_timeseries.csv')
df
```

```python
df = df.melt(id_vars=['Date','Day'], var_name='Country', value_name='Cases')
df
```

```python
df.tail(50)
```

```python
df = df.dropna()
df = df.reset_index()
df
```

```python
dc = []
dd = []
for i in range(0,738):
    j = df['Country'][i]
    if j[0] == 'C':
        c = df['Cases'][i]
        d = '-'
    elif j[0] == 'D':
        d = df['Cases'][i]
        c = '-'
    dc.append(c)
    dd.append(d)
```

```python
df['Cases'] = dc
df['Deaths'] = dd
df
```

```python
da = []
for i in range(0,738):
    j = df['Country'][i]
    if j[0] == 'C':
        c = df['Country'][i][6:]
    elif j[0] == 'D':
        c = df['Country'][i][7:]
    da.append(c)   
    
df['Country'] = da
df
```
