```python
import pandas as pd

df =pd.read_csv('./assignments/country_timeseries.csv')
df
```

```python
df = df.melt(id_vars=['Date','Day'], var_name='Country', value_name='Counts')
df = df.dropna()
df = df.reset_index()
df
```

```python

filt_c = df['Country'].str.contains('Cases', na=False)
filt_d = df['Country'].str.contains('Deaths', na=False)
df['Cases'] = df.loc[filt_c,'Counts']
df['Deaths'] = df.loc[filt_d,'Counts']
```

```python
df[['cdc','country']] = df['Country'].str.split('_',expand=True)
df = df.drop(columns=['cdc','Country'])
df.rename(columns={'country':'Country'},inplace=True)
df
```

```python
df = df.drop(columns=['Counts','index'])
df['Date']= pd.to_datetime(df['Date']) 
df = df[['Date','Day','Country','Cases','Deaths']]
```
